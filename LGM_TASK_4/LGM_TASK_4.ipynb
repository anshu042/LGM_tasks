Task : Image to Pencil Sketch with Python.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/mumbai.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (7,7))
plt.imshow(img)
plt.title("Orignal Image")
plt.show()
```

```python
img_gray = cv2.cvtColor (img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (7,7))
plt.imshow(img_gray, cmap = "gray")
plt.title("GrayScale Image")
plt.show()
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize = (7,7))
plt.imshow(img_invert, cmap = "gray")
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (23,23), sigmaX =0, sigmaY =0)
plt.figure(figsize = (7,7))
plt.imshow(img_smoothing, cmap = "gray")
plt.title("Blur Image")
plt.axis("off")
plt.show()
```

```python
final_img = cv2.divide(img_gray, 255-img_smoothing, scale = 255)
plt.figure(figsize = (7,7))
plt.imshow(final_img, cmap = "gray")
plt.title("Final Sketch Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Inverted Image")

plt.subplot(1,5,2)
plt.imshow(img_gray, cmap = "gray")
plt.axis("off")
plt.title("GrayScale Image")

plt.subplot(1,5,3)
plt.imshow(img_invert, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")

plt.subplot(1,5,4)
plt.imshow(img_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Blur Image")

plt.subplot(1,5,5)
plt.imshow(final_img, cmap ="gray")
plt.axis("off")
plt.title("Final Sketch Image")

plt.show()
```
